```python
import numpy as np
import matplotlib.pylab as plt
from colors import *
```

```python
num = 30
s = 0.2
colors_list = []
for l in [0.4,0.5,0.6]:
    plt.figure(figsize=[5,5])
    colors = getRGBAs(s,l,num)
    colors_list.append(colors)
    cnt = 0
    for i in range(int(np.sqrt(num))):
    #     print(colors[cnt,:])
        for j in range(int(np.sqrt(num))):
            plt.plot([i],[j],marker='.',markersize=50,color=colors[cnt,:])
            cnt += 1
            if cnt>num:
                break
```

```python
colors =  np.vstack(colors_list)
c = colors[np.linspace(0,colors.shape[0]-1,30,dtype=int),0:3]
step = 13
for i in range(30):
    print("{:.6f},{:.6f},{:.6f};".format(c[((i*step)%num)%30,0],c[((i*step)%num)%30,1],c[((i*step)%num)%30,2]))
```
